### Concatenação das planilhas

```python
mun_apic = pd.read_csv('/mun_apicultura.csv')
mun_apic.head(3)
```

```python
mun_caprino = pd.read_csv('/mun_caprino.csv')
mun_caprino.head(3)
```

```python
mun_frango = pd.read_csv('/mun_frango.csv')
mun_frango.head(3)
```

```python
mun_ovino = pd.read_csv('/mun_ovino.csv')
mun_ovino.head(3)
```

```python
mun_suino = pd.read_csv('/mun_suino.csv')
mun_suino.head(3)
```

```python
mun_pecuaria_bahia = pd.concat([mun_apic, mun_caprino, mun_frango, mun_ovino, mun_suino], ignore_index = True)
```

```python
mun_pecuaria_bahia.head()
```

```python
mun_pecuaria_bahia.shape
```

```python
mun_pecuaria_bahia = mun_pecuaria_bahia.drop(columns = ['Unnamed: 0'], axis = 1)
```

```python
mun_pecuaria_bahia.head(3)
```

```python
mun_pecuaria_bahia.to_csv('/mun_pecuaria_bahia.csv')
```
