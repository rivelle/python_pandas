#### Carregando e ajustandos os dataframes para a concatenação

```python
# Carregando a primeira planilha que será utilizada no merge

df1 = pd.read_csv('/seucaminho_no_drive/mun_caprino.csv')

df1.head(3)
```

```python
# Em algumas situações serão necessários alguns ajustes nas planilhas antes da realização do Merge
# Neste exemplo será necessária a remoção da coluna Unnamed

df1 = df1.drop(columns = ['Unnamed: 0'], axis = 1)
```

```python
df1.head(3)
```

```python
df1.shape
```

```python
# Carregando a segunda planilha que será utilizada no merge

df2 = pd.read_csv('/seucaminho_no_drive/mun_frango.csv')

df2.head()
```

```python
# Assim como no primeiro dataframe, aqui também será necessária a remoção da coluna Unnamed

df2 = df2.drop(columns = ['Unnamed: 0'], axis = 1)
```

```python
df2.head(3)
```

```python
df2.shape
```

#### Concatenando os arquivos com o .concat

O concat irá "empilhar" um dataframe sobre o outro respectivamente na ordem em que forem colocados na linha de comando. Neste exemplo o df1 será "empilhado" sobre o df2.

O argumento ignore_index dirá a forma como os índices serão ou não reorganizados na concatenação. Caso seu valor seja True ele irá resetar os indices mantendo uma única contagem à partir da concatenação. Caso seja False ele manterá os indices originais de cada dataframe.

```python
df_geral = pd.concat([df1, df2], ignore_index=True)

display(df_geral)
```

```python
# Após a concatenação o dataframe resultande será salvo no drive como um arquivo .csv

df_geral.to_csv('/seucaminho_no_drive/mun_pecuaria_bahia.csv')
```
